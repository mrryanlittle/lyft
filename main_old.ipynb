```python
# imports
import pandas as pd
import numpy as np
```

```python
# files
file1 = 'data/driver_rides.csv'
file2 = 'data/driver_sessions.csv'
```

```python
# initial file dataframes
driver_rides_df = pd.read_csv(file1)
driver_sessions_df = pd.read_csv(file2)
```

### Create DataFrame of When and Where Rides are Accepted
#### M1.1 Ride Length Total
#### M1.2 Ride length from Accept to Arrive
#### M1.3 Ride length from Arrive to Pickup
#### M1.4 Ride length from Pickup to Dropoff

```python
# pull file into a DataFrame
driver_rides_df
```

```python
# drop unneccesary columns, drop columns with NaN
# this data shows length of time in the lifecycle of each ride
# the lifecycle is - accept ride, pickup rider, dropoff rider -> Isolate the Lifecyele timestamps

driver_timestamps_df = driver_rides_df.drop(['status', 'accepted_lat', 'accepted_lng', 'pickup_lat','pickup_lng','dropoff_lat','dropoff_lng'], axis = 1).dropna()
driver_timestamps_df
```

```python
# I need to sort by 'accepted_timestamp'
driver_timestamps_df.sort_values(by='accepted_timestamp',axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
```
